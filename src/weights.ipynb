```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("output/mv_sp500/weights_mutual_information.csv", index_col="date")
df.index = pd.to_datetime(df.index)

stocks = df.columns.tolist()
colors = sns.color_palette("tab20", n_colors=len(stocks))

with plt.style.context("seaborn-v0_8-darkgrid"):
    fig, ax = plt.subplots(figsize=(14, 6))

    ax.stackplot(
        df.index,
        df[stocks].T.values,
        labels=stocks,
        colors=colors,
        alpha=0.9
    )

    ax.set_title("Portfolio Weight Evolution")
    ax.set_ylabel("Weight")
    ax.set_xlabel("Date")

    #ax.legend(
    #    loc="lower center",
    #    bbox_to_anchor=(0.8, 1.02),  # <-- above plot
    #    ncol=5,
    #    fontsize=8,
    #    frameon=False
    #)

    plt.tight_layout()
    plt.savefig(
        "output/mv_sp500/weight_evolution_MI.pdf",
        dpi=300,
        bbox_inches="tight"
    )
    plt.show()
```
